```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

sbarabarnaIdDataset =  pd.read_csv('BdSL_Combined_Dataset.csv') 
```

```python
sbarabarnaIdDataset
```

```python
#null values  remove


sbarabarnaIdDataset = sbarabarnaIdDataset.dropna()
#check null values exist
sbarabarnaIdDataset.isnull().sum()


```

```python
sbarabarnaIdDataset
```

```python
# add gesture_id column for identyfying unique Label আ -> 0, ই -> 1, etc.
sbarabarnaIdDataset['gesture_id'] = sbarabarnaIdDataset['label'].astype('category').cat.codes
```

```python
# Feature Engineering Function
def extract_features(df_group):
    features = {}
    for col in ['p1', 'p2', 'p3', 'p4', 'p5', 'r1', 'r2', 'r3', 'r4', 'r5']:
        features[col + '_mean'] = df_group[col].mean()
        features[col + '_std'] = df_group[col].std()
        features[col + '_max'] = df_group[col].max()
        features[col + '_min'] = df_group[col].min()
    return features
```

```python
feature_list = []
label_list = []

for gid, group in sbarabarnaIdDataset.groupby("gesture_id"):
    features = extract_features(group)
    features["label"] = group["label"].iloc[0]
    feature_list.append(features)

df_features = pd.DataFrame(feature_list)
df_features.head(100)
df_features.shape
```

```python
df_features
```

```python
# Simple SVM Implementation
import pickle

# Prepare data
X = df_features.drop('label', axis=1)
y = df_features['label']

print(f"Dataset: {X.shape[0]} samples, {X.shape[1]} features")
print(f"Classes: {y.unique()}")

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train SVM
svm_model = SVC(kernel='rbf', random_state=42)
svm_model.fit(X_scaled, y)

# Test on training data (since dataset is small)
predictions = svm_model.predict(X_scaled)
accuracy = accuracy_score(y, predictions)

print(f"\nTraining Accuracy: {accuracy:.4f}")
print(f"Correct predictions: {sum(y == predictions)}/{len(y)}")

# Save model
with open('best_svm_model.pkl', 'wb') as f:
    pickle.dump(svm_model, f)
    
with open('feature_scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

print("Model saved successfully!")
```
